# Part A: CONCEPTUAL FOUNDATION (THEORY)

## 1. What is Regularization in Machine Learning? Why is it needed?

Regularization is a technique used to reduce model complexity and prevent overfitting in machine learning. It adds a penalty term to the loss function so the model does not learn unnecessary patterns from training data and performs better on unseen data.

---

## 2. Difference between Ridge Regression (L2) and Lasso Regression (L1)

Ridge Regression uses L2 regularization and reduces coefficient values but does not make them exactly zero. Lasso Regression uses L1 regularization and can shrink some coefficients to zero, which helps in feature selection and reducing unnecessary features.

---

## 3. What is Cross-Validation and why is it important?

Cross-validation is a model evaluation technique used to measure how well a machine learning model generalizes to unseen data. It improves model reliability by testing the model on multiple subsets of the dataset and helps reduce overfitting.

---

## 4. Explain the following cross-validation techniques

### K-Fold Cross-Validation

The dataset is divided into K equal parts. The model is trained on K-1 folds and tested on the remaining fold. This process repeats K times, and the average performance is calculated.

### Stratified K-Fold Cross-Validation

This method maintains the same distribution of target values in every fold. In regression, target values are first converted into bins to preserve balanced data distribution across folds.

### Leave-One-Out Cross-Validation (LOOCV)

In LOOCV, one sample is used for testing while the remaining samples are used for training. This process repeats for every data point in the dataset and provides highly accurate evaluation.

### Time Series Split

Time Series Split is used for sequential or time-based data. It preserves the order of data by training on past data and testing on future data, avoiding data leakage.

---

## 5. Why are tree-based models less sensitive to feature scaling?

Tree-based models split data based on feature values and conditions instead of distance calculations. Because of this, scaling or normalization of features does not significantly affect the performance of Decision Trees or Random Forest models.

# PART D: CROSS-VALIDATION STRATEGIES


```python
from sklearn.model_selection import (KFold, StratifiedKFold, LeaveOneOut, TimeSeriesSplit, cross_val_score)
```

```python
# Use best tuned Ridge model for cross-validation
model = ridge_grid.best_estimator_
```

### 13. K-Fold Cross-Validation


```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

kfold_scores = cross_val_score(
    model,
    X_train_scaled,
    y_train,
    cv=kfold,
    scoring='neg_mean_squared_error'
)

kfold_mse = -kfold_scores.mean()

print("K-Fold Cross Validation MSE :", kfold_mse)
```

```python
# Stratified K-Fold Cross-Validation
# Target is binned so StratifiedKFold can balance each fold
y_binned = pd.cut(y_train.reset_index(drop=True), bins=5, labels=False)

stratified_kfold = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

X_train_scaled_reset = X_train_scaled.reset_index(drop=True)

stratified_scores = cross_val_score(
    model,
    X_train_scaled_reset,
    y_train.reset_index(drop=True),
    cv=stratified_kfold.split(X_train_scaled_reset, y_binned),
    scoring='neg_mean_squared_error'
)

stratified_mse = -stratified_scores.mean()

print("Stratified K-Fold MSE :", stratified_mse)
```

```python
# Leave-One-Out Cross-Validation (LOOCV)
loo = LeaveOneOut()

loo_scores = cross_val_score(
    model,
    X_train_scaled,
    y_train,
    cv=loo,
    scoring='neg_mean_squared_error'
)

loo_mse = -loo_scores.mean()

print("LOOCV MSE :", loo_mse)
```

```python
# Time Series Split Cross-Validation
# Pipeline used to prevent data leakage — scaler is fit only on training folds
from sklearn.pipeline import Pipeline

df_sorted = df.sort_values(by='sale_year')

X_time = df_sorted.drop(['house_price_inr', 'property_id'], axis=1)
y_time = df_sorted['house_price_inr']

tscv = TimeSeriesSplit(n_splits=5)

time_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge(alpha=ridge_grid.best_params_['alpha']))
])

time_scores = cross_val_score(
    time_pipe,
    X_time,
    y_time,
    cv=tscv,
    scoring='neg_mean_squared_error'
)

time_mse = -time_scores.mean()

print("Time Series Split MSE :", time_mse)
```

### 14. Compare All CV Strategies


```python
cv_results = pd.DataFrame({
    'Cross Validation Method': [
        'K-Fold',
        'Stratified K-Fold',
        'LOOCV',
        'Time Series Split'
    ],
    'MSE Score': [
        kfold_mse,
        stratified_mse,
        loo_mse,
        time_mse
    ]
})

print("\nCross Validation Comparison")
print(cv_results)
```

### Interpretation

> All cross-validation methods produce similar MSE scores, showing stable model performance.\
> K-Fold and Stratified K-Fold give the lowest errors.\
> Time Series Split has a slightly higher error because it respects time order, which is a stricter evaluation.

### 24. Identify signs of overfitting or underfitting in each model.



**Logic:**

- Train Error << Validation Error → Overfitting
- Train Error ≈ Validation Error → Good Fit
- Both Errors High → Underfitting


```python
models_check = {

'Ridge':[ridge_train_mse, best_ridge_mse],
'Lasso':[lasso_train_mse, best_lasso_mse],
'Decision Tree':[dt_train_mse,dt_test_mse],
'Random Forest':[rf_train_mse,rf_test_mse],
'SVR':[linear_train_mse,best_error]
}

for model_name, errors in models_check.items():
    train = errors[0]
    test = errors[1]
    print("\n", model_name)
    print("Train Error :", train)
    print("Validation Error :", test)

    # Handle edge case: if train error is 0, it is clear overfitting
    if train == 0:
        print("Overfitting (Train error is zero — model memorized training data)")
    elif test > 0 and train < (test * 0.5):
        print("Overfitting")
    elif test > 0 and abs(train - test) / test < 0.15:
        print("Good Fit")
    else:
        print("Underfitting")
```

## Interpretation:

- Ridge and Lasso show good generalization — training and validation errors are very close.
- Decision Tree clearly overfits — training error is zero (memorized all data) but validation error is much higher.
- Random Forest shows slight overfitting but still generalizes well due to bagging.
- SVR shows poor overall performance on this dataset.
